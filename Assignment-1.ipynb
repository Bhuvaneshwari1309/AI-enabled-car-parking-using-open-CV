```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('House Price India.csv')

plt.hist(df['living area'],bins=10)
plt.xlabel('living area')
plt.ylabel('Frequncy')
plt.show()

plt.scatter(df['number of bedrooms'], df['Price'])
plt.title('Scatter plot of Bedrooms vs Price')
plt.xlabel('number of bedrooms')
plt.ylabel('Price')
plt.show()

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
db = fig.add_subplot(111, projection='3d')

db.scatter(df['grade of the house'], df['Area of the house(excluding basement)'], df['Area of the basement'])
db.set_xlabel('grade of the house')
db.set_ylabel('Area of the house(excluding basement)')
db.set_zlabel('Area of the basement')
plt.show()

df.describe()

```
